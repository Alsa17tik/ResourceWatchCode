```python
import pandas as pd
import json
from operator import itemgetter
from functools import reduce
```

```python
with open('data/corpus.txt', 'r') as f:
    corpus = json.loads(f.read())

with open('data/corpus_count.txt', 'r') as f:
    corpus_count = json.loads(f.read())
    
def sort_corpus(corpus_count):
    sorted_corpus = sorted(corpus_count.items(), key=itemgetter(1))
    return list(reversed(sorted_corpus))
```

```python
sorted_corpus = sort_corpus(corpus_count)
sorted_corpus
```

```python
with open('data/nlp_results.txt', 'r') as f:
    nlp_results = json.loads(f.read())

def sort_corpuses(nlp_results):
    keys = list(nlp_results)
    sorted_counts = map(sort_corpus, nlp_results.values())
    return dict(zip(keys, sorted_counts))
```

```python
sorted_corpuses = sort_corpuses(nlp_results)
sorted_corpuses
```

```python
top_20 = pd.read_csv('data/top_20_most_connected_countries.csv')
for relation in top_20.iterrows():
    data = relation[1]
    cntry1 = json.loads(data['"c1"'])['country']
    cntry2 = json.loads(data['"c2"'])['country']
    shared_nouns = json.loads(data['"r"'])['shared_nouns']
    
    print('{} nouns shared between {} and {}'.format(shared_nouns, cntry1, cntry2))
    print()
    print('{} nouns: {}'.format(cntry1, sorted_corpuses[cntry1]))
    print()
    print('{} nouns: {}'.format(cntry2, sorted_corpuses[cntry2]))
    print()
```
